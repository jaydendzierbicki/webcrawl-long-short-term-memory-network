```python
import getpass
from sqlalchemy import create_engine  
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, TimeSeriesSplit, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.wrappers.scikit_learn import KerasRegressor
import tensorflow as tf


# Function to load data into mySQL
def extract_data_mySQL(database_name, table_name):
    user = 'root'

    # Prompt the user for a password
    password = getpass.getpass("Enter your MySQL password: ")
    host = 'localhost'
    port = 3306
    database = database_name
    engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

    # Write the DataFrame to a SQL table
    table_name =  table_name
    query = f"SELECT * FROM {table_name}"
    
    # Read query to dataframe
    df = pd.read_sql(query, engine)
    return(df)

    # Close the connection
    engine.dispose()
```

```python
data = extract_data_mySQL('MA5851_A3', 'ltsm_data_frame')
data.set_index('date', inplace=True) # Index date
data
```

```python
# Sort the data by date
data.sort_index(inplace=True)

# Set the seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Extract x and y for model 1 and model 2
y_price_prediction = data['close']
x_price_prediction = data.drop(columns=['close'])

# Split the data into train and test sets
train_size = int(len(x_price_prediction) * 0.8)
x_train, x_test = x_price_prediction[:train_size], x_price_prediction[train_size:]
y_train, y_test = y_price_prediction[:train_size], y_price_prediction[train_size:]

x_test_naive = x_test['open']

# Calculate the naive predictions
y_pred_naive = x_test_naive

# Calculate the naive MSE
mse_naive = mean_squared_error(y_test, y_pred_naive)

# Print the naive MSE
print(f'Naive Model MSE: {mse_naive}')

# Get the corresponding date index for the test set
date_index = data.index[train_size:]

# Plot the true values and predictions for the naive model
plt.figure(figsize=(12, 6))
plt.plot(date_index, y_test, label='True Values')
plt.plot(date_index, y_pred_naive, label='Naive Predictions')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Extract the closing prices
y_price_prediction = data['close']

# Use yesterday's closing price as the predicted value for today's closing price
y_pred_naive = y_price_prediction.shift(1)

# Drop the first row of the predictions, since it will be NaN due to the shift
y_pred_naive = y_pred_naive[1:]

# Get the corresponding date index for the test set
date_index = y_price_prediction.index[y_pred_naive.index]

# Calculate the naive MSE
mse_naive = mean_squared_error(y_price_prediction[1:], y_pred_naive)

# Print the naive MSE
print(f'Naive Model MSE: {mse_naive}')

# Plot the true values and predictions for the naive model
plt.figure(figsize=(12, 6))
plt.plot(date_index, y_price_prediction[1:], label='True Values')
plt.plot(date_index, y_pred_naive, label='Naive Predictions')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

```python
# Scale the data - scale sepertly to avoid data leak
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Reshape the data to fit the LSTM input shape
x_train_reshaped = x_train_scaled.reshape(x_train_scaled.shape[0], x_train_scaled.shape[1], 1)
x_test_reshaped = x_test_scaled.reshape(x_test_scaled.shape[0], x_test_scaled.shape[1], 1)
```

```python
# Set the seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create the LSTM model
def create_model(neurons=50, dropout_rate=0.2):
    model = Sequential()
    model.add(LSTM(neurons, input_shape=(x_train_reshaped.shape[1], 1), return_sequences=True))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(neurons, return_sequences=False))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Perform grid search to select the best hyperparameters
model = KerasRegressor(build_fn=create_model, verbose=0)

param_grid = {
    'neurons': [50],
    'dropout_rate': [0.2],
    'batch_size': [32, 64],
    'epochs': [100]
}

tscv = TimeSeriesSplit(n_splits=3)
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=4,
                                   cv=tscv, scoring='neg_mean_squared_error', verbose=2,
                                   random_state=42, n_jobs=-1)  # Set seed for RandomizedSearchCV/use all avaible cores

random_search_result = random_search.fit(x_train_reshaped, y_train)

# Train and evaluate the model using the best hyperparameters
best_model = random_search_result.best_estimator_
history = best_model.fit(x_train_reshaped, y_train, validation_data=(x_test_reshaped, y_test), epochs=100, batch_size=64)
y_pred = best_model.predict(x_test_reshaped)

# Print performance metrics
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Best Hyperparameters: {random_search_result.best_params_}')

# Get the corresponding date index for the test set
date_index = data.index[train_size:]

# Plot the true values and predictions
plt.figure(figsize=(12, 6))
plt.plot(date_index, y_test, label='True Values')
plt.plot(date_index, y_pred, label='Predicted Values')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Plot the training and validation loss against the epoch
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
